```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
fy15_html_path = "Resources/FY15_Data_Clean.xlsx"

fy15_df = pd.read_excel(fy15_html_path)
fy15_df.head()
```

```python
fy16_html_path = "Resources/FY16_Data_Clean.xlsx"

fy16_df = pd.read_excel(fy16_html_path)
fy16_df.head()
```

```python
fy17_html_path = "Resources/FY17_Data_Clean.xlsx"

fy17_df = pd.read_excel(fy17_html_path)
fy17_df.head()
```

```python
fy18_html_path = "Resources/FY18_Data_Clean.xlsx"

fy18_df = pd.read_excel(fy18_html_path)
fy18_df.head()
```

```python
# Combined the two dataframes by concat (puts them on tops of one another)

frames = [fy15_df, fy16_df, fy17_df, fy18_df]
fy_combined_df = pd.concat(frames, keys=['FY15','FY16','FY17','FY18'])

fy_combined_df.head()
```

```python
fy_combined_df['DECISION_DATE_YEAR'] = pd.DatetimeIndex(fy_combined_df['DECISION_DATE']).year
fy_combined_df['DECISION_DATE_MONTH'] = pd.DatetimeIndex(fy_combined_df['DECISION_DATE']).month
fy_combined_df.head()
```

```python
fy_combined_df['CASE_RECEIVED_DATE_YEAR'] = pd.DatetimeIndex(fy_combined_df['CASE_RECEIVED_DATE']).year
fy_combined_df['CASE_RECEIVED_DATE_MONTH'] = pd.DatetimeIndex(fy_combined_df['CASE_RECEIVED_DATE']).month
fy_combined_df.head()
```

```python
fy_combined_df['CASE_NUMBER'].count()
```

```python
# made sure float types were still there
fy_combined_df.dtypes
```

```python
# exported as excel and CSV
fy_combined_df.to_excel("Resources/FY_CombinedData_Clean.xlsx")
```

```python
fy_combined_df.to_csv("Resources/FY_CombinedData_Clean.csv")
```

```python
connection_string = "postgres:postgres@localhost:5432/ETL_Combined"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
fy_combined_df.to_sql(name='fy_combined', con=engine, if_exists='append', index=True)
```

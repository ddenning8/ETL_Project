```python
fy17_df2 = fy17_df[['CASE_NUMBER','DECISION_DATE','CASE_STATUS','CASE_RECEIVED_DATE','EMPLOYER_NAME','EMPLOYER_CITY',
                   'EMPLOYER_STATE','EMPLOYER_POSTAL_CODE','EMPLOYER_NUM_EMPLOYEES','AGENT_FIRM_NAME','AGENT_CITY',
                   'AGENT_STATE','PW_SOC_CODE','PW_SOC_TITLE','PW_AMOUNT_9089','WAGE_OFFER_FROM_9089','WAGE_OFFER_UNIT_OF_PAY_9089',
                   'JOB_INFO_WORK_CITY','JOB_INFO_WORK_STATE','JOB_INFO_WORK_POSTAL_CODE','JOB_INFO_JOB_TITLE',
                   'JOB_INFO_EDUCATION','JOB_INFO_EDUCATION_OTHER','FOREIGN_WORKER_INFO_CITY','FOREIGN_WORKER_INFO_STATE',
                   'FW_INFO_POSTAL_CODE','COUNTRY_OF_CITIZENSHIP','FW_INFO_BIRTH_COUNTRY','CLASS_OF_ADMISSION','FOREIGN_WORKER_INFO_EDUCATION','FW_INFO_EDUCATION_OTHER']]

fy17_df2.head()
```

```python
# The wage columns are objects need to convert to floats
fy17_df2.dtypes
```

```python
# removed comma from wage numbers but kept decimal
cols = ['PW_AMOUNT_9089','WAGE_OFFER_FROM_9089']

fy17_df2[cols] = fy17_df2[cols].replace({',': ''}, regex=True)
```

```python
# Converted the two wage columns to floats
fy17_df2['PW_AMOUNT_9089'] = pd.to_numeric(fy17_df2['PW_AMOUNT_9089'])
fy17_df2['WAGE_OFFER_FROM_9089'] = pd.to_numeric(fy17_df2['WAGE_OFFER_FROM_9089'])
```

```python
# Checked to make sure they were changed to floats
fy17_df2.dtypes
```
